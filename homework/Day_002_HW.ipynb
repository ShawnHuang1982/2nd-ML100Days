##### 符合特定值

```python
filter = (app_train["CODE_GENDER"] == "F")
df7 = app_train[filter]
df7
```

##### 符合特定值 + 範圍

```python
df8 = df7[3:5]
df8
```

```python
df6 = app_train['SK_ID_CURR']
df6
```

##### 看某一列的資料

```python
df2 = app_train.iloc[[44]]
df2
```
